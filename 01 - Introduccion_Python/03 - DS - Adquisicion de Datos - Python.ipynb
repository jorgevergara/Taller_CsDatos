### Sitios Web

Es bastante com칰n, la necesidad de extraer informaci칩n de un sitio web y de a su vez de forma autom치tica. Para ello existen dos t칠cnicas bien conocidas que son: (i) Web Scarping y (ii) Web Crawling.

#### 쯈u칠 es web Scraping?
Es el acto de procesar un documento web y extraer informaci칩n del mismo.

#### 쯈u칠 es web Crawling?
Es el proceso de encontrar y recuperar "web links", desde una lista inicial de URLs, navegando el contenido de la lista inicial y aplicando t칠cnicas de web scraping para encontrar nuevos links.

Existen varias herramientas que implementan estas t칠cnicas, una de las m치s populares es la libreria de Python [Scrapy](https://scrapy.org/).

#### Ejemplo

A continuaci칩n vamos a ver un ejemplo de como extraer informaci칩n del popular sitio web de rese침as de peliculas, [IMDb](https://www.imdb.com). Para ello vamos a usar **Python** y las librerias [Requests](https://pypi.org/project/requests/) y [BeautifulSoup](https://pypi.org/project/beautifulsoup4/), para recuperar los primeros resultados de b칰squeda de peliculas cuya fecha de lanzamiento se encuentra entre el a침o 2018 y 2019.

Si te sientes curioso, hacer el scrapping completo de un sitio web puede ser bastante m치s complicado que esto, incluso si queremos navegar en paginas hijas y descargar un sitio completo. Para ello suelen utilizarse librer칤as m치s elaboradas como [Scrapy](https://docs.scrapy.org/en/latest/intro/tutorial.html).

````python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup

# Define URL a descargar y hace la petici칩n HTTP
url = 'http://www.imdb.com/search/title?count=100&release_date=2018,2019&title_type=feature'
response = requests.get(url)

# Crea objeto BeautifulSoup para procesar HTML
soup = BeautifulSoup(response.text, 'html.parser')

# Obtengo todos los elementos H3 con clase de estilos lister-item-header (yo se que all칤 estan los t칤tulos)
elements = soup.findAll('h3', {'class': 'lister-item-header'})

# El resultado es una lista, itero para cada uno
for element in elements:
    # El titulo est치 dentro de un link
    title = element.find("a", recursive=False).next
    print(title)
````

Notar, la complejidad que implica hacer web scarping para un data scientist
- Necesitamos conocer la estructura de la p치gina web.
- Conocer las b치sicas de HTML, CSS, entre otros.
- Si la estructura de la p치gina cambia, necesito cambiar el programa que hace scraping.

### APIs REST
Muchas veces en un proyecto de ciencia de datos, tenemos el requerimiento de acceder a los datos a trav칠s de una API REST. Naturalmente no vamos a re-implementar la rueda y desarrollar un programa que establezca una conexi칩n HTTP, parse la respuesta, etc. Existen librerias que implementan clientes rest y tambi칠n librer칤as que nos ayudan a parsear respuestas en formato JSON, XML, etc. Veamos un ejemplo en el que utilizamos las librerias ```httr``` y ```jsonlite``` para invocar la API de la popular plataforma de compra venta de bienes y servicios [Mercado Libre](https://developers.mercadolibre.com.uy/es_ar). En el mismo, vamos a recuperar y mostrar informaci칩n de las categor칤as de productos que se ofrecen a trav칠s de la plataforma.

**Consumiendo servicio REST**

````python
# Importo librerias
import requests
import pandas as pd
import json

# Direcci칩n del servicio en la API REST
url = "https://api.mercadolibre.com/sites/MLC"

# Invocaci칩n al m칠todo
response = requests.get(url)

# Obtenemos el status code de la respuesta con el operador '$'
response.status_code
````

**Parseando contenido de la respuesta**

````python
# Extraigo el contenido de la respuesta como JSON
content = response.json()

# El resultado tiene varios atributos, me quedo solo con el atributo categorias
categories = content['categories']

# Construyo un data frame a partir del la lista de categorias
df = pd.DataFrame.from_dict(categories)

# Veamos el resultado
df.head(5)
````

**Parseando como JSON**

````python
# Obtengo el identificador de la categoria en el resultado, para ello selecciona la primer fila usando el
# operador iloc (https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.iloc.html)
category_id = df.iloc[0]['id']

# Direcci칩n del servicio en la API REST
url = "https://api.mercadolibre.com/categories/" + category_id

# Invocaci칩n al m칠todo
response = requests.get(url)

# Obtengo el resultado de la respuesta como JSON
content = response.json()

print(json.dumps(content, indent=4))
````

Existen ciertos desaf칤os que siempre estan presentes en la etapa de adquisici칩n de datos, los principales son:

Aveces gastamos una cantidad considerable de tiempo pre-procesando los datos para poder ingerirlos y empezar la etapa de comprensi칩n y an치lisis.
